### Lesson 10: Set

A set is an unordered collection of unique elements.
Properties:
1. No duplicates
2. Unordered  ** The order in the set is not gauranteed **
3. Mutable
4. Fast lookup

Memory map:
* list = []
* tuple = ()
* set ={}

```python
# Creating a set
numbers = {1, 2, 3}
numbers
```

```python
# Duplicate removal
nums = {10, 20, 30, 10, 20}
print(nums)
```

```python
# Creating an empty set
s = set() # empty {} represents dictionary.
```

```python
# Adding elements to the set
nums = {10, 20}
nums.add(30)
nums
```

```python
# Adding multiple elements in the set
nums.update([40, 50, 60])
nums
```

```python
#removing elements
nums.remove(50)
nums
```

```python
# pop() -> removes a random element.
value = nums.pop()
value

# Sets are unordered so it give a random value.
```

```python
# Set operations.
A = {1, 2, 3}
B = {4, 5, 6}

# Union
print(A | B) # or A.union(B)

# Intersection
print(A & B) # or A.intersection(B)

# Difference
print(A - B)
```

### Lesson 11: Dictionary

A dictionary is a collection of: Key : Value pairs.

dictionaries are created either by
1. dictionary = {} // Simple brackets
2. dictionary = dict() // dict function.

<b>NOTE</b><br>
Keys must be unique in the dictionaries.

```python
student = {
    "name": "Chitti",
    "age": 21,
    "cgpa": 7.77
}
student
```

```python
student = {
    "name": "Chitti",
    "name": "Sai"
}
student # last value win when there is a duplicate.
```

```python
student = {
    "name": "Chitti",
    "age": 21
}
```

```python
# Accessing values from dictionary.
print(student["name"])
```

```python
# it can be accessed using get() -> safe.
print(student.get("name"))
```

```python
student.get("cgpa", 0) # throws a default value if there is no key with that name.
```

Just like the functions in lists most of the functions work in dictionary also.

```python
# returning the key value pairs
student.items()
```

```python
# loop through keys
for key in student:
    print(key)
```

```python
# loop through values:
for values in student.values():
    print(values)
```

```python
# loop through key value at a time:
for key, value in student.items():
    print(key, value)
```

```python
# Nested dictionaries
students = {
    101: {
        "name": "Chitti",
        "cgpa": 7.77
    },
    102: {
        "name": "Sai",
        "cgpa": 8.2
    }
}

# Pls go through this. Since mostly we might use it and would be confusing if not go through this carefully in the later parts.
```

```python
print(students[101]["name"])
```

```python
# Merge two dictionaries
d1 = {
    "a": 1
}

d2 = {
    "b": 2
}

d1.update(d2)
d1
```
