### Reshaping, stacking, Squeezing and unSqueezing

```python
#Reshape the tensor

x = torch.arange(1, 10, 1)
x, x.shape
```

```python
x_reshaped = x.reshape(1, 1, 9)
x_reshaped, x_reshaped.shape
```

```python
#Change the view  (Changing the z changes the x also)

z = x.view(1, 9)
z, z.shape
```

```python
#Stack tensors on top of the tensors

#dim = 0 or -2 Gives the stacked tensors in the top
#dim = 1 or -1Creates same col as the row and stacks everything
#dim = -1

x_stacked = torch.stack([x, x, x, x], dim=-2)
x_stacked
```

```python
#Squeeze the x

x_reshaped.shape
```

```python
x_reshaped.shape, x_reshaped.squeeze().shape # remove the 1d

x_squeezed = x_reshaped.squeeze()
```

```python
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
x_unsqueezed
```

```python
# torch.permute - rearranges the elements of the tensors elements

x = torch.randn(2, 3, 5)
x.size()
```

```python
x.permute(2, 0, 1).shape
```
