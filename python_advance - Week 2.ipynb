### Lesson 2: Exception Handling

An Exception is a runtime error that interrupts normal program execution.
eg : 10 / 0 throws a zeroDivisionError.

Basic Syntax<br>
try:<br>
>risky_code<br>

except:<br>
>error_handling_code

```python
try:
    num = int(input("Enter a number: "))  # Enter 0 to fire the exception.
    print(100 / num)
except:
    print("Something went wrong.")
```

Working principle.

* try block starts.
* checks any error occurred in the try block execution.
* if YES except block runs
* if NO skip the except block.

A block can raise multiple exceptions.
eg: If we have to read a number and divide it by something. There we have
1. ValueError -> other than number
2. ZeroDivisionError -> denominator is zero.

```python
try:
    num = int(input("Enter a number: "))
    print(100 / num)
except ValueError:
    print("Invalid number.") # Give an char to raise this error.
except ZeroDivisionError:
    print("You can't divide by zero.") # Give 0 to raise this error.
```

Exception Object.

This prints the actual reason for why the exception has raised.

```python
try:
    print(10 / 0)
except ZeroDivisionError as e:
    print(e)
```

```python
# Catching multiple Exceptions together.
try:
    nums = int(input("Enter a number: "))
    print(100 / nums)
except (ValueError, ZeroDivisionError) as e:
    print(e)
```

```python
# else
try:
    nums = int(input())
except ValueError:
    print("Invalid number.")
else:
    print("Valid number.")
```

ABOVE OBSERVATION

We know that else can't be written independently without an if statement. But we didn't get any error in the above case.

This states that except blocks is acting like if statement.
Workflow:
* try block executes.
* if there is an error except block executes.
* If no else block executes.

<b>Note</b> Else block executes only when except block doesn't execute.

```python
try:
    nums = int(input("Enter a number: "))
except ValueError:
    print("Invalid number.")
finally:
    print("Closing.") # No matter what finally executes atlast.
```

Common built-in Exceptions
1. ZeroDivisionError
2. ValueError
3. TypeError
4. IndexError
5. KeyError
6. FileNotFoundError
7. AttributeError

What if we want to write a custom exception

This is where raise comes in...


```python
age = -5

if age < 0:
    raise ValueError("Age cannot be negative")
```

### Lesson 3: File Handling.

A file is a collection of data stored on secondary storage.
Eg: notes.txt, students.csv, photo.jpg etc

Most file handling revolves around four operations.
1. Open
2. Read
3. Write
4. Close

```python
# Opening a file uses a method open() => file = open(filename, mode)

try:
    file = open("student.txt", "r")
except FileNotFoundError as e:
    print(e)
```

Files Modes
1. r  = Read
2. w  = Write
3. a  = Append
4. x  = Create New File
5. rb = Read Binary
6. wb = Write Binary

```python
# Create a file name student.txt and write some data into it.
try:
    file = open("student.txt", "r")
    data = file.read() # reads the entire content in the file. returns str
    print(data)
except FileNotFoundError as e:
    print(e)
finally:
    file.close()
    print("Closing.")

# Remember to close the file always.
```

read() -> reads all the data from the file<br>
readline() -> reads the data line by line in the file. one line per run.<br>
readlines() -> reads the entire data line by line in one run.<br>

```python
file = open("student.txt", "r")

data = file.readlines()

print(data)

file.close()
```

```python
try:
    file = open("student.txt", "w") # writes into the file. Note: When 'w' is used the data in file is completely flushed and writes the data newly.
    file.write("chitti")
except FileNotFoundError as e:
    print(e)
finally:
    file.close()
```

```python
# writing multiple lines in a file
file = open("student.txt", "w")

file.write("Chitti\n")
file.write("Sai\n")
file.write("Ram\n")

file.close()
```

```python
# Append mode
try:
    file = open("student.txt", "a")
    file.write("Chitti by append mode\n")
except FileNotFoundError as e:
    print(e)
finally:
    file.close()
```

Constructor with parameters:<br>

A parameterized constructor in python is a special method that passed arguments to a class when creating an object.

Unlike the dynamically given parameters, These parameters are common in all the objects created by the class.

```python
class Student:

    def __init__(self, name, age):
        self.name = name
        self.age = age

s1 = Student("Ramesh", 20)
print(s1.name)
```

```python
class Student:

    name = "Ramesh"
    age = 40
    def __init__(self, name, age: int):

        if not isinstance(age, int):
            raise TypeError("Age must be an integer.")

        # Python dynamically allocates the data type so without the above two lines, it won't throw any sort of data type mismatch error.

        self.name = name
        self.age = age

s1 = Student("Raju", 40)
s2 = Student("Ravi", "kiran")

print(s1.name)
print(s2.age)
```

Instance Variable vs Class Variable.

Suppose if a class contains students names like<br>
>class Student:<br>
>>def __init__(self, name):<br>
>>>self.name = name<br>

the name is called as instance Variable. Since it changes for objects.

But class variable is a global variable in class that all the objects that are instantiated with the class has the same value.

eg:
>class student:
>>college_name = "IIIT DWD"

The college name is same throughout the class objects.

```python
class Student:

    clg_name = "IIIT DWD"

    def __init__(self, name):
        self.name = name

s1 = Student("Ravi")
s2 = Student("Raju")

print(s1.clg_name)
print(s2.clg_name) # Different objects but same class variable value.
```

Class Method

A class method is a specialized method in python that is bound to the class itself rather than individual object instances

This can do anything in the class other than changing the instance variables.

```python
class student:

    college = "ABC"

    @classmethod
    def change_method(cls, name):
        cls.college = name

student.change_method("IIIT DWD")
print(student.college) # Since it is there from the beginning before the object is created, It doesn't need any object creation also. It can be called just by the Class name.
```

Just like self refers to the Object instance, cls refers to the cls instance.

#### Lesson 6.5: Encapsulation

Encapsulation means bundling data and methods together while controlling access to the data.

```python
class Student:

    def __init__(self):
        self._cgpa = 7.77

    @property
    def cgpa(self):
        return self._cgpa

    @cgpa.setter
    def cgpa(self, value):

        if value < 0:
            raise ValueError

        self._cgpa = value
```

### Lesson 5.7: Inheritance

Inheritance is the mechanism by which one class acquires the properties and methods of another class.

Syntax:

class Parent: <br>
>pass<br>

class child(Parent):<br>
>pass

```python
class Person:

    def __init__(self, name):
        self.name = name
    def display(self):
        print(self.name)

class Student(Person):
    pass

s = Student("Chitti") # student automatically inherits __init__() and display()
s.display()
```

```python
# Adding child features.
class Person:
    def __init__(self, name):
        self.name = name
class Student(Person):
    def __init__(self, name, cgpa):
        super().__init__(name)
        self.cgpa = cgpa
s = Student("Chitti", 7.77)
print(s.__dict__)
```

What does super() do ?

super() gives access to the parent class.<br>
eg: super().__init__(name) == Person.__init__(self, name)

Advantages:
1. Cleaner
2. Supports multiple inheritance.
3. Easier maintenance.

`Method Overriding`:

Child class replaces the parent's implementation.

```python
class Person:
    def work(self):
        print("Working")
class Student(Person):
    def work(self):
        print("Studying") # child method is overriding the parent's method.

s = Student()
s.work()

class Student(Person):
    def work(self):
        super().work()
        print("Studying")

s = Student()
s.work()
```

Types of Inheritance.

1. Single Inheritance. Person <-- Student. A parent class and it is inherited to a child class.
2. Multilevel Inheritance. Person <-- Student <-- GraduateStudent. A parent is inherited to a child and the child is inherited to its child. Result it in the inheritance of parent to grandchild.
3. Hierarchical Inheritance. Person <-- Student. Person <-- Teacher. A single parent class is inherited to muliple child classes.
4. Multiple Inheritance. Parent , Parent2 <-- Child. A parents are inherited to a single child.

```python
# Single Inheritance:
class Student(Person):
    pass
```

```python
# MultiLevel Inheritance:
class GraduateStudent(Student):
    pass
```

Method Resolution Order

```python
class A:
    def show(self):
        print("A")
class B:
    def show(self):
        print("B")
class C(A, B):
    pass

c = C()
c.show()
print(C.mro())
```

```python
class A:
    def show(self):
        print("A")
class B(A):
    def show(self):
        print("B")
class C(A):
    def show(self):
        print("C")
class D(C, B):
    pass

d = D()
d.show()
print(D.__mro__) # Order says D C B A. Therefore it first takes the OBject from C.
```

### Lesson 5.8: Polymorphism

One interface, many implementations.

Instead of writing different code for different objects, we write one piece of code that works with many object types

```python
class Dog:
    def sound(self):
        print("Bow")
class Cat:
    def sound(self):
        print("Meow")
animals = [Dog(), Cat()]
for animal in animals:
    animal.sound()
```

Types of Polymorphism.

Python mainly supports:
* Method Overriding
* Duck Typing


Duck Typing

One of Python's most important concepts.

Python follows:

"If it walks like a duck and quacks like a duck, treat it as a duck."

Python doesn't care about the object's type.

It only cares whether the required method exists.

```python
class Duck:
    def speak(self):
        print("Duck")
class Person:
    def speak(self):
        print("HI")
def talk(obj):
    obj.speak()

talk(Duck())
talk(Person()) # NO obj is created. Python found the method and run it.
```

```python
# Creating Custom Operator Overloading.
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def __add__(self, other):
        return Point(
            self.x + other.x,
            self.y + other.y
        )
p1 = Point(2, 3)
p2 = Point(4, 5)

p3 = p1 + p2
print(p3.__dict__)
```
