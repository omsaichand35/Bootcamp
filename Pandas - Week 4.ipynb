### Lesson 5: loc vs iloc

loc: labels

iloc: integer positions

```python
df = pd.DataFrame(
    {
        "age": [20, 21, 19],
        "cgpa": [8.2, 7.8, 9.1]
    },
    index=["STU101", "STU102", "STU103"]
)
df
```

```python
# loc: label based selection
df.loc["STU101"]
```

```python
# multiple rows and cols
df.loc["STU102", "age"]
```

Syntax:

df.loc[row_labels, column_labels] -> value at the column_label

```python
df.loc[["STU101", "STU102"], ["age", "cgpa"]]
```

iloc: Position based selection.

```python
df.iloc[0]
```

```python
df.iloc[1, 1]
```

syntax: df.iloc[row_positions, column_positions]

```python
# slicing
df.iloc[0:2] # 0 - include, 2 - exclude.
```

```python
df.loc["STU101": "STU103"] # includes both.
```

```python
# masking.
df[df["cgpa"] > 8.0] # just like in numpy &, | are used to concat more than 1 statement.
```

### Lesson 6: Query


```python
# instead of using & or | for the statements use query() for better readablity.
df.query("cgpa > 8.0 and age < 30")
```

```python
min_cgpa = 7.3
df.query("cgpa > @min_cgpa") # @ is used to access the values outside the dataframe.
```

### Lesson 7: Membership Filtering with isin()

```python
df = pd.DataFrame({
    "name": ["Arun", "Bala", "Charan", "Divya"],
    "department": ["CSE", "ECE", "CSE", "MECH"],
    "cgpa": [8.2, 7.8, 9.1, 8.0]
})
```

```python
df[
    df["department"].isin(["CSE", "ECE"])
]
```

```python
df[df["cgpa"].between(1, 2)] # fetchs values between them.
```

```python
df.sort_values("cgpa", ascending=False)
```

```python
df.sort_values(
    by=["cgpa", "department"],
    ascending=[True, False]
)
```

```python
filtered = df[df["cgpa"] < 9.0]
filtered.index
```

```python
filtered.reset_index()
```

```python
filtered.reset_index(drop=True)
```

### Lesson 8: Missing values in the dataset.

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "name": ["Arun", "Bala", "Charan", "Divya", "Esha"],
    "age": [20, None, 19, 22, None],
    "cgpa": [8.2, 7.8, np.nan, 8.7, 8.5],
    "department": ["CSE", "ECE", None, "CSE", "ECE"]
})

df
```

```python
# detect missing values
df.isna()
# False - if not missing
# True - if missing
```

```python
df.isnull()  # isna and isnull are exactly identical methods.
```

```python
# count missing values per column
df.isnull().sum() # think of it True -> 1 and False -> 0
```

```python
df.isna().sum().sum() # returns -> total missing values.
```

```python
df.isna().mean() * 100 # percentage of missing values.
```

```python
missing_report = pd.DataFrame({
    "missing_count": df.isna().sum(),
    "missin_percent": df.isna().mean() * 100,
})
missing_report
```

```python
missing = df.isna().sum()
missing[missing > 0]
```

```python
df.dropna() # removes the missing null values
```

```python
df.dropna(how="any")
```

```python
df.dropna(how="all")
```
