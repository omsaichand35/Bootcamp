Array Properties:

Array properties are among the most frequently used attributes in ML.

1. shape - return the shape of the array. -> (in tuples)
2. ndim - returns the dimension of the array.
3. size - returns the size of the array (rows * columns)
4. dtype - This tells how each element in the array is stored.
5. itemsize - This tell how much space does each element occupies in the array.
6. nbytes - total memory used by the array.

```python
arr = np.array([[1, 2, 3],
                [3, 4, 5],
                [6, 7, 8]])
#shape
print(f"shape of the array: {arr.shape}")
#ndim
print(f"Dimensions of the array: {arr.ndim}")
#size
print(f"Number of elements in the array: {arr.size}")
#dtype
print(f"Elements datatype in the array: {arr.dtype}")
#itemsize
print(f"Size of each element in the array: {arr.itemsize}")
#nbytes
print(f"Total space occupied by the array: {arr.nbytes}")
```

```python
# Chaning datatypes
arr = arr.astype(np.float64) # np.float64 converts the values to floating points at 64 bit precision.
print(arr)
```

```python
# Explicit dtype.
arr = np.array([1, 2, 3], dtype=np.float64) # sets the datatype during the creation of the array.
print(arr)
```

`Importance of Shape in ML`

The most common reason is the matrix multiplication. All the neural network models are just matrix multiplications inside. The simple rule in the matrix multiplication is.

Given a matrix M x N and P x T in order to multiply N == T. if not we raise a shape mismatch error.

So it is compulsory to check the shape of the arrays before proceeding the futher steps.

### Lesson 6: Broadcasting.

Broadcasting in NumPy is a powerful mechanism that allows arithmetic operations to be performed on arrays of different shapes.

```python
marks = np.array([10, 20, 30, 40, 50, 60, 70, 80, 90])
marks += 5
marks # No need of loops
```

```python
a = np.array([1, 2, 3])
b = np.array([2, 3, 4])
a + b
```

```python
a = np.array([
    [1,2,3],
    [4,5,6]
])
b = np.array([10,20,30])
a + b
```

BroadCasting Rules:

1. If the shape of the arrays are same. No problem it executes.
2. If the shape of the array are not same like (3, 4) + (1, 4)  numpy streches the array to become (3, 4)
3. Missing Dimensions are assumed to be 1.

<b>The numpy starts from the rightmost dimension. Always.</b>


Note: `@` is used for matrix multiplication.

```python
a = np.array([
    [1, 2, 3],
    [2, 3, 4]
])
a.shape
```

* Scalars can be broadcast across arrays.
* Smaller arrays can behave like larger arrays without copying data.
* Broadcasting compares shapes from right to left.
* A dimension is compatible if it is equal or 1.
* Broadcasting powers image normalization, feature scaling, neural network bias addition, and many other ML operations.

### Helper Functions:

```python
# 1. Min-Max Normalization
import numpy as np
def min_max_scale(x):
    """
    Scale values between 0 and 1
    Formula is x - min(x) / (max(x) - min(x))
    """
    x = np.asarray(x)
    return (x - np.min(x)) / (np.max(x) - np.min(x))
```

```python
# 2. Standardization
import numpy as np
def standardize(x):
    """
    mean = 0
    std = 1
    Formula is (x - mean) / (std + eps)
    """
    x = np.asarray(x)
    return ((x - np.mean(x)) / np.std(x))
```

```python
# 3. One-Hot Encoding
def one_hot_encoding(labels, num_classes=None):
    labels = np.asarray(labels)
    if num_classes is None:
        num_classes = np.max(labels) + 1

    result = np.zeros((len(labels), num_classes))
    result[np.arange(len(labels)), labels] = 1
    return result
```

```python
# 4. Accuracy Scores:
def accuracy(y_true, y_pred):
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)

    return np.mean(y_true == y_pred)
```

```python
# 6. Confusion Matrix
def confusion_matrix(y_true, y_pred):
    classes = np.unique(
        np.concatenate([y_true, y_pred])
    )
    matrix = np.zeros(
        (len(classes), len(classes)),
        dtype=np.int
    )

    for t, p in zip(y_true, y_pred):
        matrix[t, p] += 1

    return matrix
```

```python
# 7. Cosine Similarity
def cosine_similarity(a, b):
    a = np.asarray(a)
    b = np.asarray(b)

    return (
        np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
    )
```

```python
# 8. Sigmod
def sigmod(x):
    return 1 / (1 + np.exp(-x))
```

```python
# 9. Relu
def relu(x):
    return np.maximum(0, x)
```

```python
# 10. Softmax
def softmax(x):
    x = np.asarray(x)
    x = x - np.max(x)
    exp = np.exp(x)
    return exp / np.sum(exp)
```

```python
# 11. Mean Squared Error
def mse(y_true, y_pred):
    return (
        np.mean((y_true - y_pred) ** 2)
    )
```

```python
# 12. Binary Cross Entropy
def binary_cross_entropy(y_true, y_pred):
    eps = 1e-15

    y_pred = np.clip(
        y_pred,
        eps,
        1 - eps
    )

    return -np.mean(
        y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred)
    )
```
